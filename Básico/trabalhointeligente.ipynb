```python
df = pd.read_csv("compras.csv")
df.columns = [col.lower().replace("ç", "c").replace("ã", "a") for col in df.columns]

con = duckdb.connect()

query = """
SELECT canal_aquisicao, COUNT(canal_aquisicao) AS total
FROM df
GROUP BY canal_aquisicao
"""

resultado = con.execute(query).fetchdf()
print(resultado)

resultado.to_excel("resultado.xlsx", index=False)
```

```python
# 1. Qual o produto mais vendido?
dados_compra['produtos_adquiridos'] = dados_compra['produtos_adquiridos'].dropna().str.replace(' ', '')
```

```python
todos_produtos = []
for produtos in dados_compra['produtos_adquiridos']:
    todos_produtos.extend(produtos.split(','))
```

```python
todos_produtos
```

```python
contagem_produtos = Counter(todos_produtos)
contagem_produtos
```

```python
produto_mais_vendido = contagem_produtos.most_common(1)[0]
produto_mais_vendido
```

```python
dados_produtos = pd.DataFrame(contagem_produtos.items(), columns=['Produto', 'Quantidade']).sort_values(by='Quantidade', ascending=False)
dados_produtos
```
